```python
def hitung_luas(panjang, lebar):
    def kali(a, b):
        return a * b
    return kali(panjang, lebar)

print("Luas 5 x 3 =", hitung_luas(5, 3))
print("Luas 10 x 7 =", hitung_luas(10, 7))
```

```python
def kuadrat(n):
    return n**2

def tambah_kuadrat(a, b):
    return kuadrat(a) + kuadrat(b)

print(tambah_kuadrat(4, 3))
```

```python
def calc_bmi(weight, height): #weight in kg, height in m
   index = weight / (height ** 2)
   return index

def check_bmi(weight, height):
   index = calc_bmi(weight, height)
   if index <= 18.5:
       print('Your BMI:', index, '- Underweight!')
   elif index > 18.5 and index <= 25:
       print('Your BMI:', index, '- Normal weight!')
   else:
       print('Your BMI:', index, '- Overweight!')

weight = float(input('Enter weight (in kg):'))
height = float(input('Enter height (in m):'))
check_bmi(weight, height)
```
