```python
from random import randint

nomor_menang = 1212
angka_sekarang = randint(1200, 1220)
print('Nomor tiket lotre kamu adalah:', angka_sekarang)

if angka_sekarang == nomor_menang:
    print('Anda menang, selamat!')
else:
    print('Semoga lain kali lebih beruntung!')
```

```python
jackpot = 12345
guess = (input("Mulai lotre (enter - start, off - stop)"))
while guess != 'off':
  lot = random.randint(12340,12350)
  print(lot)

  if lot != jackpot:
    print("Nomor Salah")
  else:
    print("Selamat anda menang lotre")
    break

  guess = (input("Mulai lotre (enter - start, off - stop)"))
```

```python
start = time()
ketik = input('Tulis suatu kalimat atau paragraf: ')
stop = time()

duration = round(stop - start, 2)
print(f'Lama mengetik: {duration} detik')
size = len(ketik)
print(f'Jumlah karakter: {size} karakter')

cpm = round(size/(duration/60))
print(f'Kecepatan mengetik: {cpm} karakter per menit')
```
