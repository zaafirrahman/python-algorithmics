```python
class Car():
    def __init__(self, brand, color, price):
        self.brand = brand
        self.color = color
        self.price = price

    def print_info(self):
        print(self.brand, 'adalah mobil keren')
        print(self.brand, 'berwarna', self.color)
        print('Harga mobil tersebut', self.price, 'dolar')

rolls_royce = Car('RR','hitam',15000)
rolls_royce.print_info()

avanza = Car('Avanza', 'merah', 1000)
avanza.print_info()

porsche = Car('Porsche','blue', 1000000)
porsche.print_info()

audi = Car('Audi', 'biru', 100000)
audi.print_info()
```

```python
from time import sleep

class Hero():
   def __init__(self, name, health, power, weapon):
       self.name = name
       self.health = health
       self.power = power
       self.weapon = weapon

   def print_info(self):
       print('Menyapa tokoh ->', self.name )
       print('Tingkat kesehatan:', self.health)
       print('Kekuatan attack:', self.power)
       print('Jenis senjata:', self.weapon, '\n')

class Warrior(Hero):
   def serang(self, musuh):
       sleep(3)
       print(self.name, 'menyerang', musuh.name, 'menggunakan', self.weapon)
       musuh.health -= self.power
       print('Health Point', musuh.name, musuh.health, '\n')

alpha = Warrior('Alpha', 1000, 200, 'Shotgun')
alpha.print_info()

gamma = Warrior('Gamma', 1000, 180, 'Pedang')
gamma.print_info()
print('---------------------------')


print('-> PERTARUNGAN DIMULAI! ')
while alpha.health > 0 or gamma.health > 0:
  alpha.serang(gamma)
  if gamma.health <= 0:
    print('Alpha MENANG')
    break
  gamma.serang(alpha)
  if alpha.health <= 0:
    print('Gamma MENANG')
    break
```

```python
class Hero():
   def __init__(self, name, health, armor, power, weapon):
       self.name = name #string
       self.health = health #angka
       self.armor = armor #angka
       self.power = power #angka
       self.weapon = weapon #string

   def print_info(self):
       print('Menyapa tokoh ->', self.name )
       print('Tingkat kesehatan:', self.health)
       print('Level pertahanan:', self.armor)
       print('Kekuatan attack:', self.power)
       print('Jenis senjata:', self.weapon, '\n')

   def strike(self, enemy):
       print('-> STRIKE! ')
       print(self.name , 'menyerang', enemy.name , 'menggunakan', self.weapon, '\n')
       enemy.armor -= self.power
       if enemy.armor < 0:
           enemy.health += enemy.armor
           enemy.armor = 0
       print(enemy.name , 'jatuh.')
       print('class armor drop ke', enemy.armor)
       print('Tingkat kesehatan drop ke', enemy.health, '\n')

class Warrior(Hero):
    def hello(self):
        print('-> PAHLAWAN BARU! Seorang prajurit terampil muncul dari kedalaman hutan ', self.name )



hiro = Hero('Hiro', 240, 789, 70, 'Shield')
hiro.print_info()
print('-----------------------------')

biru = Hero('Biru', 380, 245, 189, 'Sword')
biru.print_info()
print('-----------------------------')

hiro.strike(biru)
biru.strike(hiro)
hiro.strike(biru)
biru.strike(hiro)
```

```python
class Animal():         # Superclass
    def eat(self):
        print("Animal eats")
    def sleep(self):
        print("Animal sleeps")

class Hewandarat(Animal):    # Subclass
    def walks(self):
        print("animal walks")

dog = Hewandarat()
dog.eat()
dog.sleep()
dog.walks()

class Hewanair(Animal):
    def swim(self):
        print("animal swim")

fish = Hewanair()
fish.eat()
fish.sleep()
fish.swim()
```
