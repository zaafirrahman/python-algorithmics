```python
grup = []

orang = input('Masukkan nama (0 - stop): ')

while orang != '0':
   grup.append(orang)
   orang = input('Masukkan nama (0 - stop): ')

print('Grup telah dibuat:', grup)
```

```python

grup2  = ['ADMIN']
peserta = input('Masukkan nama (0 - stop): ')
while peserta != '0':
  if peserta in grup2:
      print('Peserta sudah terdaftar')
  else:
      grup2.append(peserta)
  peserta = input('Masukkan nama (0 - stop): ')

grup2.sort()
print('Grup telah diketik:', grup2)
```

```python
results = [225, 220, 199, 263, 259, 225, 226]
average = 0
maks = 0

for i in results:
  average += i
  if maks < i:
      maks = i

average = average / len(results)
print('Hasil:', results)
print('Hasil rata-rata:', average)
print('Hasil maksimum:', maks)
```

```python
#Mencari informasi list

results2 = [225, 220, 199, 263, 259, 225, 226]

average2 = sum(results2) / len(results2)
max2 = min(results2)

print('Hasil:', results2)
print('Hasil rata-rata:', average2)
print('Hasil minimum:', max2)
```

```python
data = [20, 80, 40, 30]
print(max(data))
print(min(data))
```

```python
#Memisah dan menggabung list

angka = input('Masukkan angka (pisahkan dengan spasi): ')
angka = angka.split(' ')
print(angka)
print('-'.join(angka))
```

```python
#Menjumlah dan mengalikan list (seperti string)

contoh1 = [1, 2, 3, 4, 5]
contoh2 = [6, 7, 8, 9, 10]

print(contoh1 + contoh2)
print(contoh1 * 3)
```

```python
#Mengoperasikan nilai iterable dalam list

score = ['34', '22', '79', '81', '50', '71']
for i in range(len(score)):
   score[i] = int(score[i])
print(score)
```

```python
score2 = ['34', '22', '79', '81', '50', '71']
score_new = list(map(int, score2))
print(score_new)
```
