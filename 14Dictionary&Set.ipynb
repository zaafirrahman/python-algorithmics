```python
barang = {'HP':10,
          'Laptop':5,
          'Tablet':7,
          'Headset':16
          }

print(barang)

cek_keys = barang.keys()
cek_values = barang.values()

print(cek_keys)
print(cek_values)

print(barang['Headset'])
```

```python
# Berapa pasang barang yang tersedia?
cek_barang = len(barang)
print('Jumlah jenis barang:',cek_barang)

# Berapa stok barang tertentu?
nama = input('Masukkan nama barang: ')
cek_stok = barang[nama]
print('Stok tersedia:',cek_stok)
```

```python
# Menambahkan pasangan key value
barang['Keyboard'] = 5
print(barang)

# Menghapus pasangan
del barang['Headset']
print(barang)
```

```python
# Contoh program

print('In stock:', barang.keys())
wish = input('Masukkan barang yang diinginkan: ')
if wish in barang:
  print('Jumlah stok tersedia:', barang[wish])
  take = input('Apakah Anda ingin membeli? (ya/tidak) ')
  if take == 'ya':
      print('Pembelian dikonfirmasi')
      barang[wish] -= 1
      print('Jumlah stok tersisa:', barang[wish])
else:
  print('Barang tidak ditemukan')
```

```python
your_songs = {
    'Is It The Answer':'Reality Club',
    'Nina':'.Feast',
    'Kita ke Sana':'Hindia',
    'Terbuang Dalam Waktu':'Barasuara',
    'Secukupnya':'Hindia',
    'Gala Bunga Matahari':'Sal Priadi'
}
```

```python
my_set = set(my_songs.values())
your_set = set(your_songs.values())

print(my_set)
print(your_set)
```

```python
print('Daftar seluruh artist:', my_set.union(your_set))
print('Artist favorit keduanya:', my_set.intersection(your_set))
print('Saya tidak menyukai:', your_set.difference(my_set))
print('Kamu tidak menyukai:', my_set.difference(your_set))
```
