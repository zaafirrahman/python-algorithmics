```python
while True:
    try:
        children_amount = int(input('Masukkan jumlah anak '))
        sweets_amount = int(input('Masukkan jumlah permen '))
        break
    except:
        print('Error! Angka harus berupa bilangan bulat')

try:
    portion = sweets_amount/children_amount
    print('Setiap anak akan menerima', portion, 'permen')
except:
    print('Kesalahan pembagian! Tidak bisa membagi dengan 0')
```
