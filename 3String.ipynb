```python
# Mencari panjang string

feedback1 = input('Masukkan kata: ')
panjang = len(feedback1) #length

print('Panjang dari string adalah: ', panjang)
```

```python
# Mengekstrak substring (slicing)

feedback2 = 'aku sedang makan'
nama = feedback2[11]

print(nama)
```

```python
# Mengekstrak substring (slicing)

feedback = 'aku sedang makan'
nama = feedback[4:9]

print(nama)
```

```python
# Mencari kata tertentu

feedback3 = 'saya bangun pagi setiap hari'
cari = feedback3.find('Pagi')

print(cari)
```

```python
# Mengubah string menjadi lowercase

feedback4 = 'Besok adalah Hari Minggu dan besok Saya libur'
feedback4_low = feedback4.lower()

#print(feedback4_low)

cari_low = feedback4_low.find('minggu')

print(cari_low)
```

```python
ulasan = input('Masukkan ulasan:')

ulasan_low = ulasan.lower()
negatif = ulasan_low.find('tida nyaman')

print(negatif)
```
