```python
# Mencari panjang string

feedback1 = 'Haloo'
panjang = len(feedback1) #length

print('Panjang dari string adalah: ', panjang)
```

```python
# Mengekstrak substring (slicing)

feedback2 = 'Halo saya Karina dan saya senang belajar Python di Algonova'
nama = feedback2[9]

print(nama)
```

```python
# Mengekstrak substring (slicing)

feedback = 'Halo saya Zaky dan saya senang belajar Python di Algonova'
nama = feedback[10:14]

print(nama)
```

```python
# Mencari kata tertentu

feedback3 = 'Halo saya Rayan dan saya senang belajar python di Algonova'
cari = feedback3.find('roblox')

print(cari)
```

```python
# Mengubah string menjadi lowercase

feedback4 = 'Halo saya Angeline dan saya senang belajar Python di Algonova'
feedback4_low = feedback4.lower()

print(feedback4_low)

cari_low = feedback4_low.find('python')

print(cari_low)
```

```python
ulasan = input('Masukkan ulasan:')

ulasan_low = ulasan.lower()
negatif = ulasan_low.find('tida nyaman')

print(negatif)
```
