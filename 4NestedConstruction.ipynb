```python
"""
perhitungan total belanja
program ini memungkinkan pengguna untuk menghitung total belanja
tanpa kalkulator tambahan.
"""
harga1 = int(input('Masukkan harga barang 1: ')) #ini adalah harga barang pertama
harga2 = int(input('Masukkan harga barang 2: '))
total = harga1 + harga2 # variabel total jumlah harga barang
print('Total belanja yakni sebesar:', total)
```

```python
'''
kode berikut merupakan
perhitungan gaji
'''

# perhitungan gaji
honor = float(input("honor per jam: ")) # var honor per jam
jam = int(input("Jam kerja dalam seminggu: ")) # var jam kerja
gaji = honor * jam # hasil kali honor dengan jam
print(f'Gaji dalam seminggu: {gaji} dari honor per jam {honor} dan jam kerja {jam} jam')
```
