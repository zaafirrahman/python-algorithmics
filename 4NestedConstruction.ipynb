```python
'''
kode berikut merupakan
perhitungan gaji
'''

# perhitungan gaji
honor = float(input("honor per jam: ")) # var honor per jam
jam = int(input("Jam kerja dalam seminggu: ")) # var jam kerja
gaji = honor * jam # hasil kali honor dengan jam
print(f'Gaji dalam seminggu: {gaji} dari honor per jam {honor} dan jam kerja {jam} jam')
```
