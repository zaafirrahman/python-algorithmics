# **TIPE DATA LOGIKA (BOOLEAN)**

```python
# float
# integer
# string
# boolean
```

```python
a = int(input('a = '))
b = int(input('b = '))
print('--------------')

print('a > b ', a > b)
print('a < b ', a < b)
print('a == b ', a == b)
print('a != b ', a != b)
```

```python
pesanan = int(input('Jumlah pesanan: '))
stok = int(input('Jumlah stock: '))

print('-'*10)
print('Pesanan tersedia:', pesanan <= stok)
```

```python
c = int(input('c = '))
d = int(input('d = '))
print('--------------')

print('c > d or d > c ', c>d or d>c)
print('c > d and d > c ', c>d and d>c)
print('c > d and d < c ', c>d and d<c)
```

```python
stok = int(input('Jumlah stok yang masuk: '))
print('----------------------------')

print('Stok diizinkan masuk:', stok >= 50 and stok <= 300)
```

# **OPERATOR BERSYARAT**

```python
rasa_favorit = input('Masukkan rasa favorit Anda (misalnya, manis, pedas, gurih): ')
rasa_favorit = rasa_favorit.lower()

if rasa_favorit == 'manis':
    print('Rekomendasi makanan untuk Anda: Kue, Cokelat, atau Es Krim.')
elif rasa_favorit == 'asam':
    print('Rekomendasi makanan untuk Anda: Jeruk, Apel ijo.')
elif rasa_favorit == 'pedas':
    print('Rekomendasi makanan untuk anda: Baso mercon')

```

```python
angka = int(input('Masukkan sebuah angka: '))

if angka > 0:
    print('Angka yang Anda masukkan adalah positif.')
elif angka == 0:
    print('Angka yang Anda masukkan adalah nol.')
else:
    print('Angka yang Anda masukkan adalah negatif.')
```
