```python
# Lulus: >=70 (B), >=90 (A)
# Tidak lulus: <70 (C), <50 (D)

nilai = int(input('Masukkan nilai: '))

if nilai >= 70:
      print("Selamat, Anda luluss.")
      if nilai >= 90:
        print("Anda mendapatkan nilai A.")
      else:
        print("Anda mendapatkan nilai B.")
else:
      print("Maaf, Anda belum lulus.")
      if nilai >= 50:
        print("Anda mendapatkan nilai C.")
      else:
        print("Anda mendapatkan nilai D.")
```

```python
total = int(input('Masukkan total belanja:'))

if total >= 100:
  credit = input('Apakah anda menggunakan kredit?:').lower()
  if credit == 'ya':
    diskon = total*0.75
    print('Total belanja anda menjadi:', total - diskon)
  else:
    diskon = total*0.5
    print('Total belanja anda menjadi:', total - diskon)
else:
  print('Total belanja anda menjadi:', total)
```
