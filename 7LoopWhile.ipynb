```python
harga = int(input('Masukkan harga barang pertama (0 untuk menyelesaikan):'))
total = 0

while harga != 0:
   total += harga
   harga= int(input('Masukkan harga berikutnya (0 untuk menyelesaikan):'))

print('Harga total tanpa diskon:', total)

total *= 0.9
print('Harga total dengan diskon:', total)
```

```python
# Kalkulator Tabungan

hasil_akhir = int(input("Masukkan hasil_akhir tabungan: "))
harian = int(input("Masukkan jumlah tabungan hari ini: "))

hari = 0
tabungan = 0

while tabungan < hasil_akhir:
    hari += 1
    harian = int(input("Masukkan jumlah tabungan hari ini: "))
    tabungan += harian
    print(f'Hari ke-{hari} = {tabungan}')

print(f'Butuh {hari} hari untuk mencapai hasil_akhir {tabungan}')
```
