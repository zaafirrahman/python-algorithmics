```python
word = 'python'
print(len(word))

for i in word:
   print('halo')

print('Program Selesaii')
```

```python
biodata = ['naura', 13, 7]
print(len(biodata))

for i in biodata:
   print(i)
```

```python
password = input('Masukkan password: ')
alphabet = 'abcdefghijklmnopqrstuvwxyz'

for i in password:
   print(alphabet.find(i)+1)
```

```python
login = input('Masukkan info masuk: ')
wrong = '=?*^$№@_'

for i in login:
   if i in wrong:
       print('Simbol terlarang: ', i)
```

```python
for i in range(10): # range (stop)
  print(i)
```

```python
for i in range(100,110): # range (start, stop)
  print(i)
```

```python
for i in range(1,10,2): # range(start, stop, jump)
  print(i)
```

```python
jumlah = int(input('Jumlah anggota: '))

for i in range(jumlah):
   name = input('Masukkan nama: ')
   print('Selamat datang,', name)

print('Obrolan grup telah dibuat!')
```
