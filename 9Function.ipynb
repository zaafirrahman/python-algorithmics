```python
def hypotenuse(a,b):
  c = (a**2 + b**2)**0.5
  return c

alas = int(input("Masukkan alas: "))
tinggi = int(input("Masukkan tinggi: "))

miring = hypotenuse(alas,tinggi)

print("Panjang sisi miring adalah", miring)
```

```python
def l_rectangle(width,height):
  area = width * height
  return area

def k_rectangle(width,height):
  cir = 2 * (width + height)
  return cir

panjang = int(input("Masukkan panjang: "))
lebar = int(input("Masukkan lebar: "))

luas = l_rectangle(panjang,lebar)
keliling = k_rectangle(panjang,lebar)

print("Luas persegi panjang adalah", luas)
print("Keliling persegi panjang adalah", keliling)
```
