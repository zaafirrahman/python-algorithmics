```python
import pandas as pd
df = pd.read_csv('GoogleApps.csv')
```

```python
# Counting (value_counts)
df['Category'].value_counts()
```

```python
# Filtering + counting
df[df['Rating'] > 4.5]['Category'].value_counts()
```

```python
# Basic Groupby
df.groupby('Category').size()
```

```python
df.groupby('Type')['Rating'].mean()
```

```python
# Multi Insight from 1 group (Aggregation)
df.groupby('Type')['Price'].agg(['min', 'max', 'mean'])
```

```python
df.groupby('Type')['Rating'].agg(['min','mean','max'])
```

```python
# Perbandingan antar grup
temp = df.groupby('Type')['Rating'].mean()
print(temp['Paid'] - temp['Free'])
```

```python
# Ratio
temp = df['Content Rating'].value_counts()
print(temp['Teen'] / temp['Everyone 10+'])
```
